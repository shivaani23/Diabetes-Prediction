```python
def distance(x1,x2):
    return np.sqrt(sum((x1-x2)**2))


def knn(x,y,test,k=4):
    d=[]
    for i in range(len(x)):
        dist=distance(x[i],test)
        d.append((dist,y[i]))
    
    #print(d)
    d=np.array(sorted(d))[:,1]
    #print(d)
    d=d[:k]
    #print(d)
    t=np.unique(d,return_counts=True)
    #print(t)
    idx=np.argmax(t[1])
    '''if t[1,0]==t[1,1]:
        idx=t[0,1]'''
    #print(idx)
    pred=int(t[0][idx])
    
    return pred
```

```python
y_pred=[]
for i in range(576):
    m=knn(x_train,y_train,test=[x_train[i]])
    y_pred.append(m)
    

y_pred=np.array(y_pred)
y_pred.reshape((576,1))
print(y_pred)
df=pd.DataFrame(y_pred,columns=["Outcome"])
df.to_csv("submission.csv",index=False)
```
